```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('/content/Iris.csv')
```

```python
iris
```

```python
iris.dtypes
```

```python
plt.hist(iris['SepalLengthCm'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.hist(iris['SepalWidthCm'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
plt.hist(iris['PetalLengthCm'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Petal Length Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
plt.hist(iris['PetalWidthCm'], bins=10, edgecolor='k', alpha=0.7)
plt.xlabel('Petal Width Cm')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
species_count = iris['Species'].value_counts()
plt.bar(species_count.index, species_count.values)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
plt.show()
```
